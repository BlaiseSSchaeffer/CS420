```python
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

# Unsupervised Learning Iris

```python
iris = load_iris()
data = iris.data
```

```python
kmm = KMeans(3,n_init=30)
kmm.fit(data)
kmm_label = kmm.predict(data)
kmm_label
```

```python
zero = 0
one = 0
two = 0
for prediction in kmm_label:
    if prediction == 0:
        zero += 1
    elif prediction == 1:
        one += 1
    else:
        two += 1
        
print("Zero Predicted:", zero)
print("One Predicted:", one)  # These are most likely Setosa for the current run, it always gets all 50 of these correct (100%)
print("Two Predicted:", two)
```

```python
targets = iris.target

zero = 0
one = 0
two = 0
for known in targets:
    if known == 0:
        zero += 1
    elif known == 1:
        one += 1
    else:
        two += 1
        
print("Zero:", zero)
print("One:", one)
print("Two:", two)
```

# Image Processing

```python
from PIL import Image
im = Image.open("2013-Roster-Pict-copy.jpg")

pixels = list(im.getdata())

# width, height = im.size
# pixels = [pixels[i * width:(i + 1) * width] for i in range(height)]
```

```python
import numpy as np
picData = np.array(im)

print(picData.shape)
x = picData.reshape(500 * 333, 3)
y = x.reshape(500, 333, 3)

newIm = Image.fromarray(y)
plt.imshow(newIm)
```

```python
pic = KMeans(3,n_init=30)
pic.fit(pixels)
pic_label = pic.predict(pixels)
```

```python
print(len(pic_label))

values = []
for point in pic_label:
    if point not in values:
        values.append(point)
        
print(values)
```

```python
centers = pic.cluster_centers_
centers
```

```python
for i in range(len(x)):
    x[i] = centers[pic_label[i]]
```
